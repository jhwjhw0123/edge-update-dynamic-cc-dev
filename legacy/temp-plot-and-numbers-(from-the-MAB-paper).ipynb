```python
import os
import pickle
import numpy as np
import pysnooper
import random
import copy
import tqdm
from scipy.stats import truncnorm
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
# parameter
num_arms = 50000
# arm_setting = 'clear_cut_setting'  
arm_setting = 'mix_in_setting'
```

```python
# rst_id = 'num_multiplier='+str(trial_mult_factor)+'_exp='+str(trial_exp_factor)
read_dir = './experiments/'+arm_setting+'/num_arms='+str(num_arms)+'/'
```

```python
fig = plt.figure(figsize=(30, 7), dpi=600)

# Creating axes instance
# ax = fig.add_axes([0, 0, 1, 1])
ax_master = fig.add_subplot(111)

ax1 = fig.add_subplot(131)

img1 = mpimg.imread(read_dir+'num_multiplier=1000_exp=1.png')
# Creating plot
bp = ax1.imshow(img1)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.spines['bottom'].set_visible(False)
ax1.spines['left'].set_visible(False)
ax1.axes.get_xaxis().set_ticks([])
ax1.axes.get_yaxis().set_ticks([])
ax1.set_xlabel('T=1000K')

ax2 = fig.add_subplot(132)
img2 = mpimg.imread(read_dir+'num_multiplier=1000_exp=2.png')
# Creating plot
bp = ax2.imshow(img2)
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['bottom'].set_visible(False)
ax2.spines['left'].set_visible(False)
ax2.axes.get_xaxis().set_ticks([])
ax2.axes.get_yaxis().set_ticks([])
ax2.set_xlabel('T=1000K^2')

ax3 = fig.add_subplot(133)
img3 = mpimg.imread(read_dir+'num_multiplier=1000_exp=3.png')
# Creating plot
bp = ax3.imshow(img3)
ax3.spines['top'].set_visible(False)
ax3.spines['right'].set_visible(False)
ax3.spines['bottom'].set_visible(False)
ax3.spines['left'].set_visible(False)
ax3.axes.get_xaxis().set_ticks([])
ax3.axes.get_yaxis().set_ticks([])
ax3.set_xlabel('T=1000K^3')


ax_master.spines['top'].set_visible(False)
ax_master.spines['right'].set_visible(False)
ax_master.spines['bottom'].set_visible(False)
ax_master.spines['left'].set_visible(False)
ax_master.axes.get_xaxis().set_ticks([])
ax_master.axes.get_yaxis().set_ticks([])
ax_master.set_ylabel("Regret (log10-scale)")

# show plot
plt.show()
```

## T = 1000K

```python
uniform_exploration_regrets_linear = np.load(read_dir+'num_multiplier=1000_exp=1_uniform.npy')
naive_esp_alg_regrets_linear = np.load(read_dir+'num_multiplier=1000_exp=1_naive_eps.npy')
log_eps_alg_regrets_linear = np.load(read_dir+'num_multiplier=1000_exp=1_log_eps.npy')
loglog_eps_alg_regrets_linear = np.load(read_dir+'num_multiplier=1000_exp=1_loglog_eps.npy')
logstar_eps_alg_regrets_linear = np.load(read_dir+'num_multiplier=1000_exp=1_logstar_eps.npy')
goa_esp_alg_regrest_linear = np.load(read_dir+'num_multiplier=1000_exp=1_goa_eps.npy')
```

```python
base_line_linear = np.mean(uniform_exploration_regrets_linear)
print('Comparison of the mean regret, ')
print(np.mean(uniform_exploration_regrets_linear)/base_line_linear)
print(np.mean(naive_esp_alg_regrets_linear)/base_line_linear)
print(np.mean(log_eps_alg_regrets_linear)/base_line_linear)
print(np.mean(loglog_eps_alg_regrets_linear)/base_line_linear)
print(np.mean(logstar_eps_alg_regrets_linear)/base_line_linear)
print(np.mean(goa_esp_alg_regrest_linear)/base_line_linear)
```

```python
base_median_linear = np.median(uniform_exploration_regrets_linear)
print('Comparison of the median regret, ')
print(np.median(uniform_exploration_regrets_linear)/base_median_linear)
print(np.median(naive_esp_alg_regrets_linear)/base_median_linear)
print(np.median(log_eps_alg_regrets_linear)/base_median_linear)
print(np.median(loglog_eps_alg_regrets_linear)/base_median_linear)
print(np.median(logstar_eps_alg_regrets_linear)/base_median_linear)
print(np.median(goa_esp_alg_regrest_linear)/base_median_linear)
```
